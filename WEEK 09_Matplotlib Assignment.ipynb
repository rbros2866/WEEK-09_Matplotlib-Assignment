Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional plane.

It's particularly useful for visualizing the relationship between two variables and identifying patterns, clusters, or trends within the data.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, c='blue', marker='x', label='Data Points')

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.show() 
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

**subplot() Function:**

Creates a grid of subplots within a single figure.

Useful for displaying multiple plots in a compact layout.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

plt.figure(figsize=(12, 8))

# Subplot 1
plt.subplot(2, 2, 1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Subplot 2
plt.subplot(2, 2, 2) 
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Subplot 3
plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Subplot 4
plt.subplot(2, 2, 4) 
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.tight_layout() 
plt.show() 
```

**Q4:** What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of data visualization that represents categorical data with rectangular bars.

 It's used to compare different categories or groups by depicting the values associated with them. 
 
 Bar plots are particularly effective for visualizing the distribution, comparison, and ranking of data across categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Bar Plot - Company Profits')
plt.show()

```

```python

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.ylabel('Company')
plt.xlabel('Profit')
plt.title('Horizontal Bar Plot - Company Profits')
plt.show()
```

**Q5:** What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a statistical visualization tool used to display the distribution of a dataset. 

It presents a summary of the data's central tendency, dispersion, and potential outliers.

 The plot consists of a box that represents the interquartile range (IQR), which contains the middle 50% of the data, and "whiskers" that extend from the box to show the extent of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=labels)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
